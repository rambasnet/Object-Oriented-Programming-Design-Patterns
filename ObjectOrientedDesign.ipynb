## Object Oriented
- Objects - real-world things that can be felt, touched, sensed, manipulted, etc.
- Software objects are not tengible, but they're models of real-world objects
- object is a collection of **data** called **state** and associated **behaviors** called **methods** that act upon the data

## Object-oriented analysis (OOA)
- analyzing software design; modeling problem, task, or a system as software objects and their interactions
- part of software requirement analysis to understand what needs to be done and how
- find out what features that users/actors/stakeholders of the software need to solve the problem
- this phase of software engineering may be better called *object-oriented exploration*
- gather requirements, interview clients, understand business logic, etc.

## Object-oriented design (OOD)
- process of converting requirements into an implementation specification
- name the objects, define the behaviors, and their interactions among each other
- transforming *what* should be done into *how* 

## Object-oriented programming (OOP)
- process of converting a design into a working program that does what the product owner originally requested
- some popular development models are: **waterfall** and **iterative**
- **iterative** is more popular in modern software engineering


## Objects and Classes
- *object* is a collection of data with associated behaviors
- *class* is a kind or *type* of object
- Unified Modeling Language (UML) is commonly used to design the classes and their interactions

## Hiding details and creating the public interface (API)
- key purpose of modeling an object in OOD is to determine the public interface of that object
- interface is the collection of attributes and methods that other objects can access to interact with that object
- other objects do not need, and are not allowed, to access the internal workings of the object
- E.g., real-word example of working of TV and remote
- this technique of hiding the internal details is called **information hiding** or broadly **ecapsuluation**
- proper design of public interface is very important in software design as many other objects may depend on just those interface to interact with the objects and chaning them often may be difficult
- interface is designed with easy of use in mind (KISS principle) and NEVER ease of implementation

## Abstraction
- another key OOD term related to information hiding and encapsulation
- abstraction means dealing with the level of detail that is most appropriate to a given task
- process of extracting a public interface from inner details
    - e.g., a car's driver needs to interact with the steering wheel, accelerator, and brakes; everythin else under the hood is ignored
    - mechanic works at a different level of abstraction
- UML depicting different level of abstractions for differnet actors

## Class relationship types
- there are three types of class relationships: Association, Composition, Aggregation
- Association is the simplest of the relationship that tells how the objects are related
    - e.g., many apples go into a barrel
    - any one orange can go into exactly one basket
    - a basket can hold many oranges


## Composition
- important OOD principle; many design patterns use it
- composition is: **has-a** relationship
- act of collecting several objects together to create new one
- composition is a good choice when one object is a part of another object
    - e.g., a fossil-fueled car is composed of engine, transmission, starter, headlights, windshields, dashboard, sets, among many other parts
    - engine in turn is composed of crank shaft, valves, pistons, etc.
- an aggregate relationship is more general
- a composite relationship is also an aggregate relationship but not vice versa    

### Difference between composition and aggregation
- the lifespan of objects can help differentiate the two relationships
    - If the composite (outside) object controls when the related (inside) objects are created and destroyed, composition is most suitable.
    - If the related object is created independently of the composite object, or can outlast that object, an aggregate relationship makes more sense.

## Inheritance
- the **is-a** relationship is formed by inheritance, e.g.
    - Deep Blue is a chess player
    - Gary Kasparov is a player
- inheritance is modeled on biology
- child class inherits members of parent class similar to genetic and phoenitic inheritance in biology
- in Chess set, among 32 pieces, there are only six different types of pieces (pawns, rooks, bishops, knights, king, and queen)
- a powerful tool for extending behavior and reusing features
